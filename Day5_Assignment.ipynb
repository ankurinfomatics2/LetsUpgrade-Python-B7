### **Question 1**

Write a program to identify sub list [1,1,5] is there in the given list in the same order, if yes print “it’s a Match” if no then print “it’s Gone” in function.

Example -
* Listy =[1,5,6,4,1,2,3,5] - it’s a Match
* Listy = [1,5,6,5,1,2,3.6] - it’s Gone

```
subList = [1,1,5]
mainList = [1,5,6,4,1,2,3,5,6,7]
 
k = 0
l = 0
m = 0
 
for i in range(k, len(subList)):
  for j in range(l, len(mainList)):
    if subList[i] == mainList[j]:
      k = i + 1
      l = j + 1
      m = m + 1
      continue
 
if len(subList) == m:
  print("All items of sublist matched in main list")
else:
  print("Sublist is not matched with mainlist")
```

### **Question 2**

Make a Function for prime numbers and use Filter to filter out all the prime numbers from 1-2500

```
def primeNumber(num):
  
  count = 0
  
  for j in range (0, int(num/2)+1):
    if j != 0 and num % j == 0:
      count = count +1
  
  if count < 2:
    return True
  else:
    return False
```

```
lst = list(range(1,2500))
lst_primeNumbers = filter (primeNumber, lst)
 
print("Prime Number list:\n\n", list(lst_primeNumbers))
```

```
def  toUpper(text):
  text = text.split()

  #print (text)
  text = list(map(textToUpperLambda, text))

  text = str(" ".join(tuple(text)))

  #print(text)
  return text
```
