For example - You make a fibonacci series function,in which your input range is been defined by the decorator program input.

```
def getFibbo_Level(fib_series):
  def wrap_function():
    a = int(input("Enter the level at which you want fibonnacci series:\t"))

    fib_series(a)
  return wrap_function
```

```
@getFibbo_Level
def calc_fib_series(level):
  count = 0
  n1, n2 = 0, 1
  fib = []
  print("\nFibbonacci series is: ", end=" ")
  while count < level:
    print(n1,end=", ")
    # fib = fib.append(n1)
    nth = n1 + n2
    n1 = n2
    n2 = nth
    count += 1
  
```
