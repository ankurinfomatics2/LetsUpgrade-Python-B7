#### **Question 1**

Print the first ArmStrong number in the range of 1042000 to 702648265 and exit the loop as soon as you encounter the first armstrong number.

```
for i in range (1042000,702648265+1): 
  t = i             # as we need to retain the original number
  sum = 0
  pow = len(str(i))

  while t != 0:
    a = t%10
    sum += a**pow
    t = t//10

  if sum == i:
    print("First Armstrong number between 1042000 to 702648265 is: ", i)
    break
    
```
