### **Question 2**

##### Dictionary and its default functions

**1. Create Dictionary**

```
d = {1: "one", 2: "two"}
d
```

**2. Removes all Items**: *clear()*

```
d.clear()

print('Updated dictionary: ', d)
```

**3. Returns Shallow *Copy* of a Dictionary**: *copy()*

```
d = {1: "one", 2: "two"}
d2 = d.copy()

print("Old Dictionary: ",d)
print("New Dictionary: ",d2)
```

**4. Creates dictionary from given sequence**: *fromkeys()*

```
person = {'name', 'age', 'sex'}     #Created the list
value = 10

d3 = dict.fromkeys(person, value)   #as Dictionory should have 'key' & 'value' pair, so supplying as arguments

print("Newly formed Dictionary: ",d3)
print("Type of d3:\t\t", type(d3))
```

**5. Returns Value of The Key**: *get()*

```
person = {'name': 'Phill', 'age': 22}   #Created the dictionary

print('Name:\t', person.get('name'))    #get() returns the value of key
print('Age:\t', person.get('age')) 
```

**6. Returns view of dictionary's (key, value) pair**: *items()*

```
person = {'name': 'Phill', 'age': 22}

print(person.items())
```

**7. Returns View Object of All Keys**: *keys()*

```
person = {'name': 'Phill', 'age': 22}

print(person.keys())
```

**8. Removes and returns element having given key**: *pop()*

```
person = {'name': 'Phill', 'age': 22}
element = person.pop('name')

print('The popped element is:\t', element)
print('The dictionary is:\t', person)
```

**9. Returns & Removes Latest Element From Dictionary**: *popitem()*

```
person = {'name': 'Phill', 'age': 22, 'salary': 3500.0}

popitem = person.popitem()         # ('salary', 3500.0) is inserted at the last, so it is removed.

print('Popped Item:\t\t', popitem)
print('Updated Dictionary:\t', person)
```

**10. Add an item into the existing dictionary**

```
person = {'name': 'Phill', 'age': 22}        # Existing Dictionary

person['profession'] = 'Plumber'             # Add an item

print('Updated Dictionary:', person)
```

**11. Inserts Key With a Value if Key is not Present**: *setdefault()*

```
person = {'name': 'Phill'}

salary = person.setdefault('salary')      #this key is not in the dictionary but inserting as default

print('Updated Dictionary: ',person)
print('Value of Salary key: ',salary)
```

**12. Updates the Dictionary**: *update()*
method updates the dictionary with elements from a dictionary object or an iterable object of key/value pairs

```
d = {1: "one", 2: "three", 3:"four", 5: "five"}
d1 = {2: "two", 3: "three", 4: "four"}

d.update(d1)            # updates the value of key 2 & 3 and add key 4
print(d)
```

**13. Returns view of all values in dictionary**: *values()*
method returns a view object that displays a list of all the values in the dictionary

```
person = {'name': 'Phill', 'age': 22} 

print (person.values())
```

### **Question 3**

##### Sets and its default functions

**1. Create sets**

```
mySet = {1.0, "Hello", (1, 2, 3)}

print(mySet)
type(mySet)
```

**2. Add an element to the set**: *add()*

```
mySet = {1.0, "Hello", (1, 2, 3)}

mySet.add("orange")

print(mySet)
```

**3. Remove all the elements from the set**: *clear()*

```
fruits = {"apple", "banana", "cherry"}

fruits.clear()

print(fruits)
```

**4. Return a copy of the set**: *copy()*

```
mySet = {1.0, "Hello", (1, 2, 3)}

x = mySet.copy()

print(x)
```

**5. Return a set containing the difference between two or more sets**: *difference()*

```
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

z = x.difference(y)

print(z)
```

**6. Remove the items in this set that are also included in another, specified set**: *difference_update()*

```
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

x.difference_update(y)

print(x)
```

**7. Remove the specified item**: *discard()*

```
fruits = {"apple", "banana", "cherry"}

fruits.discard("banana")

print(fruits)
```

**8. Return a set, that is the intersection of two other sets**: *intersection()*

```
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

z = x.intersection(y)

print(z)
```

**9. Remove the items in this set that are not present in other, specified set(s)**: *intersection_update()*

```
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

x.intersection_update(y)

print(x)
```

**10. Return whether two sets have a intersection or not**: *isdisjoint()*

```
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "facebook"}

z = x.isdisjoint(y)       #Return True if no items in set x is present in set y

print(z)
```

**11. Return whether another set contains this set or not**: *issubset()*

```
x = {"a", "b", "c"}
y = {"f", "e", "d", "c", "b", "a"}

z = x.issubset(y)           #Return True if all items set x are present in set y
print(z)

z = y.issubset(x)           #Return True if all items set y are present in set x
print(z)
```

**12. 	Return whether this set contains another set or not**: *issuperset()*

```
x = {"f", "e", "d", "c", "b", "a"}
y = {"a", "b", "c"}

z = x.issuperset(y)           #Return True if all items set y are present in set x

print(z)
```

**13. Remove an element from the set**: *pop()*

```
fruits = {"apple", "banana", "cherry"}

fruits.pop()            #Remove a random item from the set

print(fruits)
```

**14. Remove the specified element**: *remove()*

```
fruits = {"apple", "banana", "cherry"}

fruits.remove("banana")         #Remove "banana" from the set

print(fruits)
```

**15. Return a set with the symmetric differences of two sets**: *symmetric_difference()*

```
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

z = x.symmetric_difference(y)     #Return a set that contains all items from both sets, except items that are present in both sets

print(z)
```

**16. inserts the symmetric differences from this set and another**: *symmetric_difference_update()*

```
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

x.symmetric_difference_update(y)      #Remove the items that are present in both sets, AND insert the items that is not present in both sets

print(x)
```

**17. Return a set containing the union of sets**: *union()*

```
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

z = x.union(y)          #Return a set that contains all items from both sets, duplicates are excluded

print(z)
```

**18. Update the set with the union of this set and others**: *update()*

```
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

x.update(y)           #Insert the items from set y into set x

print(x)
```
